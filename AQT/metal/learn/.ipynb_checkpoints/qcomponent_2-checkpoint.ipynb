```python
import qiskit_metal as metal 
from qiskit_metal import designs, draw 
from qiskit_metal import MetalGUI, Dict, open_docs
```

```python
design = designs.DesignPlanar() 
gui = MetalGUI(design) 
```

```python
from qiskit_metal.qlibrary.qubits.transmon_pocket import TransmonPocket
design.overwrite_enabled = True 
q1 = TransmonPocket(design, 'Q1') 

gui.rebuild() 
gui.autoscale() 
```

First, we'll copy the component and then manually modify its coordinates to be at (1, 0) instead of (0, 0):

```python
q1_copy = design.copy_qcomponent(q1, 'Q1_copy')
q1_copy.options.pos_x = '1.0mm'
gui.rebuild() 
gui.autoscale() 
```

```python
q1_anothercopy = design.copy_qcomponent(q1, 'Q2_copy')
q1_anothercopy.options.pos_x = '-1.0mm' 
gui.rebuild()
gui.autoscale() 
```

```python
newcopies = design.copy_multiple_qcomponents(
    [q1, q1_copy, q1_anothercopy], 
    ['Q7', 'Q8', 'Q9'], 
    [
        dict(pos_y='1.0mm'), 
        dict(pos_y='2.0mm'), 
        dict(pos_y='3.0mm')
    ]
)
gui.rebuild()
gui.autoscale()
```
