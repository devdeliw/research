```python
import gdspy
import qnldraw as qd
import numpy as np 

import matplotlib.pyplot as plt

from qnldraw import paths, shapes, components 
```

## GDSPY PolygonSets

```python
rectangle = [(0, 0), (1.5, 0), (1.5, 1), (0, 1)]
triangle = [(-0.25, 1), (1.75, 1), (0.75, 2)]

polyset1 = gdspy.PolygonSet([rectangle, triangle])
polyset2 = gdspy.PolygonSet([[(0.6, 0), (0.9, 0), (0.9, 0.5), (0.6, 0.5)]], layer=2)


fig = qd.plot([polyset1, polyset2])
```

```python
polyset1 = qd.boolean(polyset1, None, 'or', layer=1)
house = qd.boolean(polyset1, polyset2, 'not')

fig1 = qd.plot(polyset1)
fig2 = qd.plot(house)
```

## PathBase

CPW traces can be drawn with the `paths` module. Once a CPW object is created, you can draw straight segments, turns, tapers, and open gaps with the CPW methods.

```python
qd.qdParams['round.tolerance'] = 1.25

trace = paths.CPW(20, 10)
trace.open(10).segment(200).meander(5, 200, 50, turn=-1).segment(200).open(10)

trace1 = paths.CPW(20, 10)
trace1.segment(300).shift(100, radius=90).segment(300)

fig = qd.plot(trace)
fig = qd.plot(trace1)
```

```python
trace = paths.CPW.between((-500, 0), (500, 0), width=20, gap=10)
trace.open(10)

fig = qd.plot(trace, figsize=(8,2), grid=False, axis=False)
```
